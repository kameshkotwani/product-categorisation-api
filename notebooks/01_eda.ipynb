## Target Column EDA

```python
import matplotlib.pyplot as plt

# Count the occurrences of each category
category_counts = Y.value_counts()

# Plot the counts as a bar chart
category_counts.plot(kind='bar', figsize=(28, 6), color='skyblue')
plt.title('Count of Categories')
plt.xlabel('Category Name')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Insights about target column.
1. It is a category column with 129 distinct values.
2. It is positively right skwed, with Eau De Parfum having largest count, implicating a class imbalance.
